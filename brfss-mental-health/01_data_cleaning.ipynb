```python
df_clean = df_raw[['_AGEG5YR', 'SEXVAR', '_RACE', 'EDUCA', 'INCOME3', 'MARITAL', 'EMPLOY1', '_URBSTAT', 
        'GENHLTH', 'PHYSHLTH', 'ADDEPEV3', 'DIABETE4', 'CVDINFR4', 'ASTHMA3', 'HAVARTH4', 'CHCCOPD3', 
        'EXERANY2', 'SMOKDAY2', 'AVEDRNK3', 'DRNK3GE5',
        '_HLTHPL1', 'MEDCOST1', 'PERSDOC3', 
        'MENTHLTH', '_MENT14D']]
```

These don't match the 7/9 pattern:
* _AGEG5YR -- 7 and 9 are valid options (Age 50 to 54; Age 60 to 64 respectively); 14 is Don't know/Refused/Missing
* SEXVAR -- only has two options: 1/Male, 2/Female
* _RACE -- 7 is a valid option (Multiracial, non-Hispanic), 9 is Don't know/Not sure/Refused
* EMPLOY1 -- 7 is a valid option (Retired), 9 is Refused




These match the 7/9 pattern:
* EDUCA -- doesn't have a 7 but does have a 9 = Refused
* INCOME3 -- uses 77 and 99
* MARITAL -- doesn't have a 7 but does have 9 = Refused
* _URBSTAT -- just has two options (1, 2) and BLANK
* GENHLTH
* PHYSHLTH -- uses 77 and 99
* ADDEPEV3
* DIABETE4
* CVDINFR4
* ASTHMA3
* HAVARTH4

* AVEDRNK3 -- uses 77 and 99
* DRNK3GE5 -- uses 77 and 99
* _HLTHPL1 -- doesn't have a 7 but does have a 9 = Refused
* MEDCOST1
* PERSDOC3

```python
# Replacing 88/None with 0; 77/Don't know or Not sure with NaN, and 99/Refused with NaN
df_clean['MENTHLTH'] = df_clean['MENTHLTH'].replace({88:0, 77:np.nan, 99:np.nan})
```

```python
df_clean['MENTHLTH'].value_counts(dropna=False).head(10)
```

```python
df_clean['_MENT14D'].value_counts(dropna=False)
```

```python
df_clean['_MENT14D'] = df_clean['_MENT14D'].replace(9, np.nan)
```

```python
df_clean['_MENT14D'].value_counts(dropna=False)
```

```python
# Replacing 7 and 9 responses with NaN
vars_7_9 = ['EDUCA', 'MARITAL', '_URBSTAT', 'GENHLTH', 'ADDEPEV3', 'DIABETE4', 'CVDINFR4', 'ASTHMA3', 'HAVARTH4', 'CHCCOPD3', 'EXERANY2', 
            'SMOKDAY2', '_HLTHPL1', 'MEDCOST1', 'PERSDOC3']

for var in vars_7_9:
    df_clean[var] = df_clean[var].replace({7:np.nan, 9:np.nan})
```

```python
# Replacing 77 and 99 responses with NaN
vars_77_99 = ['INCOME3', 'PHYSHLTH', 'AVEDRNK3', 'DRNK3GE5']

for var in vars_77_99:
    df_clean[var] = df_clean[var].replace({77:np.nan, 99:np.nan})
```

```python
df_clean.info()
```

```python
# Replacing 14/Don't know/Refused/Missing with NaN for _AGEG5YR
df_clean['_AGEG5YR'] = df_clean['_AGEG5YR'].replace({14:np.nan})

# Replacing 9/Don't know/Not sure/Refused with NaN for _RACE
df_clean['_RACE'] = df_clean['_RACE'].replace({9:np.nan})

# Replacing 9/Refused with NaN for EMPLOY1
df_clean['EMPLOY1'] = df_clean['EMPLOY1'].replace({9:np.nan})
```

```python
df_clean.info()
```

```python
df_clean.to_csv('df_clean.csv', index=False)
```
