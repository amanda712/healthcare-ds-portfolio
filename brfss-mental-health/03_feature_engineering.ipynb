```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('df_clean.csv')
```

```python
df.shape
```

```python
df.info()
```

```python
# Missingness across all variables
df.isnull().sum().sort_values(ascending=False)
```

```python

# Drop missing values for the target variables
df = df.dropna(subset=['_MENT14D', 'MENTHLTH'])
df.shape

```

```python
# Replacing missing values with 0
df[['AVEDRNK3', 'DRNK3GE5']] = df[['AVEDRNK3', 'DRNK3GE5']].fillna(0)

df[['AVEDRNK3', 'DRNK3GE5']].isnull().sum()
```

```python
# Replacing missing values with 0 and recoding existing values from least to most smoking
df[['SMOKDAY2']] = df[['SMOKDAY2']].fillna(0)

df['SMOKDAY2'] = df['SMOKDAY2'].replace({1:3, 2:2, 3:1})

df['SMOKDAY2'].value_counts().sort_index()
```

```python
df['INCOME3'].value_counts().sort_index()
```

```python
# Creating indicator column for missings in INCOME3, converting to 1/0
df['INCOME3_missing'] = df['INCOME3'].isnull().astype(int)

df['INCOME3_missing'].value_counts()
```

```python
# Imputing missing in INCOME3 with the median
df['INCOME3'] = df['INCOME3'].fillna(df['INCOME3'].median())
```

```python
df['INCOME3'].value_counts().sort_index()
```

```python
# Replacing missings for remaining categorical variables with under 5% missings with the mode
mode_variables = ['_HLTHPL1','_RACE', '_URBSTAT', '_AGEG5YR', 'EMPLOY1', 'MARITAL', 'PERSDOC3', 
                  'ADDEPEV3', 'CVDINFR4', 'HAVARTH4', 'EDUCA', 'CHCCOPD3', 'ASTHMA3', 'MEDCOST1', 'GENHLTH', 'EXERANY2', 'DIABETE4']

for x in mode_variables:
    df[x] = df[x].fillna(df[x].mode()[0])
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['PHYSHLTH'].value_counts().sort_index()
```

```python
# Imputing missing in PHYSHLTH with the median
df['PHYSHLTH'] = df['PHYSHLTH'].fillna(df['PHYSHLTH'].median())

df['PHYSHLTH'].value_counts().sort_index()
```

```python
df.isnull().sum()
```

```python
df.to_csv('df_featured.csv', index=False)
```
