```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('df_clean.csv')
```

```python
df.shape
```

```python
df.info()
```

```python
# Missingness across all variables
df.isnull().sum().sort_values(ascending=False)
```

```python

# Drop missing values for the target variables
df = df.dropna(subset=['_MENT14D', 'MENTHLTH'])
df.shape

```

```python
# Replacing missing values with 0
df[['AVEDRNK3', 'DRNK3GE5']] = df[['AVEDRNK3', 'DRNK3GE5']].fillna(0)

df[['AVEDRNK3', 'DRNK3GE5']].isnull().sum()
```

```python
# Replacing missing values with 0 and recoding existing values from least to most smoking
df[['SMOKDAY2']] = df[['SMOKDAY2']].fillna(0)

df['SMOKDAY2'] = df['SMOKDAY2'].replace({1:3, 2:2, 3:1})

df['SMOKDAY2'].value_counts().sort_index()
```

```python
df['INCOME3'].value_counts().sort_index()
```

```python
# Creating indicator column for missings in INCOME3, converting to 1/0
df['INCOME3_missing'] = df['INCOME3'].isnull().astype(int)

df['INCOME3_missing'].value_counts()
```

```python
# Imputing missing in INCOME3 with the median
df['INCOME3'] = df['INCOME3'].fillna(df['INCOME3'].median())
```

```python
df['INCOME3'].value_counts().sort_index()
```

```python
# Replacing missings for remaining categorical variables with under 5% missings with the mode
mode_variables = ['_HLTHPL1','_RACE', '_URBSTAT', '_AGEG5YR', 'EMPLOY1', 'MARITAL', 'PERSDOC3', 
                  'ADDEPEV3', 'CVDINFR4', 'HAVARTH4', 'EDUCA', 'CHCCOPD3', 'ASTHMA3', 'MEDCOST1', 'GENHLTH', 'EXERANY2', 'DIABETE4']

for x in mode_variables:
    df[x] = df[x].fillna(df[x].mode()[0])
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['PHYSHLTH'].value_counts().sort_index()
```

```python
# Imputing missing in PHYSHLTH with the median
df['PHYSHLTH'] = df['PHYSHLTH'].fillna(df['PHYSHLTH'].median())

df['PHYSHLTH'].value_counts().sort_index()
```

```python
df.isnull().sum()
```

```python
df.to_csv('df_featured.csv', index=False)
```

```python
df = pd.get_dummies(df, columns=['_RACE', 'EMPLOY1', 'MARITAL', 'DIABETE4'], drop_first=True)
```

```python
df.shape
```

```python
df.columns.tolist()
```

`drop_first=True` dropped the first category of each variable, which becomes the reference category in the model.

* _RACE dropped _RACE_1.0 (White, non-Hispanic) — this is the reference group
* EMPLOY1 dropped EMPLOY1_1.0 (Employed for wages) — reference group
* MARITAL dropped MARITAL_1.0 (Married) — reference group
* DIABETE4 dropped DIABETE4_1.0 (Yes, diabetes) — reference group

```python
# Recoding to 0/1 for binary variables

df[['SEXVAR', '_HLTHPL1', '_URBSTAT', 'ADDEPEV3', 'ASTHMA3', 'CHCCOPD3', 'CVDINFR4', 'EXERANY2', 'HAVARTH4', 'MEDCOST1', 'PERSDOC3']] = df[['SEXVAR', '_HLTHPL1', '_URBSTAT', 'ADDEPEV3', 'ASTHMA3', 'CHCCOPD3', 'CVDINFR4', 'EXERANY2', 'HAVARTH4', 'MEDCOST1', 'PERSDOC3']].replace({2:0})
```

```python
print(df['SEXVAR'].value_counts())
print(df['ADDEPEV3'].value_counts())
```

```python
df.to_csv('df_encoded.csv', index=False)
```

```python
# Binarizing _MENT14D for the classification model
# 0 = no frequent distress (categories 1 and 2; 0 - 13 bad days)
# 1 = frequent distress (category 3; 14+ bad days)

df['MENT14D_binary'] = df['_MENT14D'].replace({1:0, 2:0, 3:1})
```
